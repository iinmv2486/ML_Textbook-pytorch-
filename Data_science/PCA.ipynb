```python
from sklearn.datasets import load_iris

irirs = load_iris()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=4)

Y = pca.fit_transform(irirs['data'])

plt.scatter(Y[:,0], Y[:,1], c=irirs['target'])
plt.show()

plt.plot(pca.explained_variance_, "-o")
plt.show()
```

# PCA via Eigen Decomposition

```python
from numpy import linalg
import matplotlib.pyplot as plt

X = irirs.data
X = X - X.mean(axis=0)

C = (X.T @ X) / X.shape[0]
L, Q = linalg.eigh(C)

print("L: ", L)
print("Q: ", Q)

plt.scatter(X @ Q[:,3], X @ Q[:,2], c=irirs.target)
plt.show()

plt.plot(sorted(L, reverse=True), "-o")
plt.show()
```
